```python
#pie chart
slices=[15,7,9,13]
courses=['python','database','android','big data']
col=['red','brown','black','blue']
plt.pie(slices,labels=courses,colors=col,explode=(0.1,0,0,0))
plt.show()
```

```python
#stack plot
num=[1,2,3,4,5]
a=[7,8,6,11,7]
b=[2,3,4,3,2]
c=[7,8,7,2,2]
d=[8,5,7,8,13]
plt.plot([],[],color='red')
plt.plot([],[],color='yellow')
plt.plot([],[],color='green')
plt.plot([],[],color='blue')
plt.stackplot(num,a,b,c,d,colors=['red','yellow','green','blue'])
plt.show()
```

```python
import numpy as np
x=np.linspace(0,10,100)
y1=x
y2=x**2
y3=x**3
y4=np.sqrt(x)
#plt.show()
#plt.figure()
plt.subplot(2,2,1)
plt.plot(x,y1,'r')
plt.title('y1=x')

plt.subplot(2,2,2)
plt.plot(x,y2,'g')
plt.title('y2=x**2')

plt.subplot(2,2,3)
plt.plot(x,y3,'b^')
plt.title('y3=x**3')
```

```python
#quiver
fig,ax=plt.subplots()
x_pos=0
y_pos=0
x_direct=2
y_direct=1
ax.quiver(x_pos,y_pos,x_direct,y_direct)
plt.show()
```

```python
fig,ax=plt.subplots()
x_pos=0
y_pos=0
x_direct=1
y_direct=1
ax.quiver(x_pos,y_pos,x_direct,y_direct,scale=5)
plt.show()
```
