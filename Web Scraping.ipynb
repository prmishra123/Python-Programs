# Web Scraping

```python
from urllib.request import urlopen as uReq
from bs4 import BeautifulSoup as soup 

quotes_page = 'http://bluelimelearning.github.io/my-fav-quotes/'
uClient = uReq(quotes_page)
page_html = uClient.read()
uClient.close()
page_soup = soup(page_html, "html.parser")
quotes = page_soup.findAll("div",{"class": "quotes"})
```

```python

for quote in  quotes:
    
    fav_quote = quote.findAll("p", {"class": "aquote"})
    aquote = fav_quote[0].text.strip()

    fav_authors = quote.findAll("p",{"class":"author"})
    author = fav_authors[0].text.strip()
    
    
    print(aquote)
    print(author)
  
```

# Scraping Flipkart

```python
from urllib.request import urlopen as uReq
from bs4 import BeautifulSoup as soup
redmi_page='https://www.flipkart.com/redmi-note-7-pro-nebula-red-128-gb/p/itmffyqg5qhzzh4z?pid=MOBFDXZ3TYYTK3UT&lid=LSTMOBFDXZ3TYYTK3UTAC9SXS&marketplace=FLIPKART&sattr[]=color&sattr[]=storage&sattr[]=ram&st=storage&otracker=AS_QueryStore_OrganicAutoSuggest_0_7'
uClient=uReq(redmi_page)
page_html=uClient.read()
uClient.close()

page_soup= soup(page_html,"html.parser")
details=page_soup.findAll("script",{"type":"application/ld+json"})#line 92
detail=details[0].text.strip()

print(detail,"\n")#details in header part
```

```python
for detail in details:
    des=detail.findAll("description")
print(des)#don't delete
```

```python
meta=page_soup.findAll("meta")
print(meta)
print("\n")
```

```python
contents=detail.findAll("content")
print(contents)#dont delete
```

```python
body=page_soup.findAll("body")
print(body,"\n")#scrap every content in the body
```

```python
print("Scraped description and comments,reviews by the customers.Details as follows:\n")
for paras in body:
    pdetails=paras.findAll("p")
    #print(len(pdetails))
for k in range(0,70):
    para=pdetails[k].text.strip()#114
    print(para,"\n")
```

```python
print("Scraped description and comments,reviews by the customers.Details as follows:\n")#with headings
quotes=page_soup.findAll("div",{"class":""})
quote=quotes[0].text.strip()
print(quote,"\n")#112 line part3
```

```python
print("Specifiation Table Fully Scrapped. Details as follows:\n")
z=1#for indexing
for detail in body:
    details=detail.findAll("td")
    #print(len(details))
    for l in range(0,126):
        aquote=details[l].text.strip()
        if l%2==0:
            
            print(z,aquote.title())#129 to be printed
            z+=1
        else:
            print(" ",aquote,"\n")
```
